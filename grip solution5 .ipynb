```python
#import the libraries
import numpy as np
from pandas_profiling import ProfileReport
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

import numpy as np
import seaborn as sns
# Loading the iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
y = iris.target
df.head(10)
```

```python
df.describe()
```

```python
df.info()
```

```python
profile=ProfileReport(df,title='Pandas Profiling Report',explorative=True)
profile.to_widgets()
```

###  define the Decision Tree Algorithm

```python
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(df,y)

print('Decision Tree Classifer Created')
```

### Let us visualize the Decision Tree to understand it better.



```python
from sklearn import tree

featName=['sepal length ( in cm)','sepal width (in cm)','petal length (in cm)','petal width (in cm)']
clsName=['Iris-Setosa','Iris-Versicolor','Iris-virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (10,6), dpi = 350)

tree.plot_tree(dtree, feature_names = featName, class_names = clsName, filled = True);
```
